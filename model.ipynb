```python
import os
import torch
```

```python
with open("text.txt", "r") as file:
    text = file.read()
```

```python
print(text[:100])
```

```python
vocab = sorted(list(set(text)))
```

```python
print("".join(vocab))
```

```python
stoi = {letter : i for i, letter in enumerate(vocab)}
itos = {i : letter for i, letter in enumerate(vocab)}
```

```python
encoder = lambda s : [stoi.get(c, 9999) for c in s]
```

```python
decoder = lambda l : "".join([itos.get(i, "") for i in l])
```
